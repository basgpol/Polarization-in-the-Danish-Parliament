## Part 1: Loading packages

```python
import requests
import json
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

## Part 2: Creating function to collect voting data from API

```python
# Saving url to voting data from API
start_url = "http://oda.ft.dk/api/Stemme?$inlinecount=allpages"
```

```python
def ft_url_collect(url, obs = None):
    """ Function that collects data from the API of the Danish Parliament
        The function takes the follwing inputs:
        
        1) URL collected from the API (e.g. "2012"). The default is year "2000".
        2) The number of observations you want to collect. Default is all observations
        """
    response = requests.get(url) # GET-request
    url_data = json.loads(response.text) #loading data as json-format
    
    # Sets the numbers of obs to max if the obs-options is not used
    if obs == None:
        obs = int(url_data["odata.count"])
    
    print("You are collecting " + str(obs) + " observations")
    
    try:
    
        #creating base df
        df = pd.DataFrame(url_data["value"])

        #calculating the number of iterations
        iterations = int(obs/100)-1 # each site has 100 observations
        print("You will do " + str(iterations) + " iterations")

        #Finding link to the second site of observations
        url_next = url_data["odata.nextLink"]

        for i in np.arange(0,iterations):
            response_iteration = requests.get(url_next) # get-request for each iterations
            url_data_iteration = json.loads(response_iteration.text) #loading data as json-format for each iteration
            df_iteration = pd.DataFrame(url_data_iteration["value"]) #creating df with obs for each iterations
            df = pd.concat([df,df_iteration]) # appending to the base df
            url_next = url_data_iteration["odata.nextLink"]
            print("Iteration " + str(i+1) + " is done. " + str(iterations-i) + " iterations left.")

        return(df) 
    
    except: #Continues if an error happens
        pass
```
